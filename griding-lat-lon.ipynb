```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from collections import defaultdict, Counter
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_path = '../../kaggleData/2sigma/'
train_df = pd.read_json(data_path+"train.json")
test_df = pd.read_json(data_path+"test.json")
train_test = pd.concat([train_df, test_df], 0)
```

```python
train_df['outline'] = False
#outlier search function from darnal
for i in ['latitude', 'longitude']:
        Q1 = train_df[i].quantile(0.005)
        Q3 = train_df[i].quantile(0.995)
        IQR = Q3 - Q1
        upper = Q3
        lower = Q1
        train_df.ix[(train_df[i]>upper)|(train_df[i]<lower),'outline'] = True
```

```python
train_df.ix[train_df['outline']==True,['latitude', 'longitude','interest_level']]
```

```python
non_out = train_df.ix[train_df['outline']==False,['latitude', 'longitude','interest_level']]
```

```python
### Rent interest graph of New-York
sns.lmplot(x="longitude", y="latitude", fit_reg=False, hue='interest_level',
           hue_order=['low', 'medium', 'high'], size=9, scatter_kws={'alpha':0.4,'s':30},
           data=non_out);
plt.xlabel('Longitude');
plt.ylabel('Latitude');
```
